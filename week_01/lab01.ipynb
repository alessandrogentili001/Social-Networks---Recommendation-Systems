### Problem 1 

Draw a square lattice without using built-in tools.

### Solution

```python
import matplotlib.pyplot as plt

def draw_square_lattice(n):
    
    # initialize the figure 
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Plot the lattice points and connections
    for i in range(n):
        for j in range(n):
            # Plot the vertices 
            ax.plot(j, i, 'bo')  # Blue points (o marker)
            
            # Plot horizontal dedges 
            if j < n - 1:
                ax.plot([j, j+1], [i, i], 'k-')  # Black lines between vertices horizontally
            
            # Plot vertical edges 
            if i < n - 1:
                ax.plot([j, j], [i, i+1], 'k-')  # Black lines between vertices vertically

    # Set grid and ticks
    ax.set_aspect('equal')  # Ensures square aspect ratio
    plt.grid(False)         # Disable the default grid lines
    plt.xticks(range(n))    # Set x-ticks to grid points
    plt.yticks(range(n))    # Set y-ticks to grid points
    plt.xlim(-0.5, n-0.5)   # Limits for x-axis
    plt.ylim(-0.5, n-0.5)   # Limits for y-axis

    plt.title(f"{n}x{n} Square Lattice")
    plt.show()

# Test the function
draw_square_lattice(5)
```

### Problem 2 

Draw a chain with your own implementation.

### Solution 

```python
import matplotlib.pyplot as plt

def draw_chain(n):
    
    # initialize the figure 
    fig, ax = plt.subplots(figsize=(6, 2))
    
    # Plot the chain vertex and edges 
    for i in range(n):
        # Plot the vertex 
        ax.plot(i, 0, 'bo')  # Blue points (o marker)
        
        # Plot horizontal edges between vertices
        if i < n - 1:
            ax.plot([i, i+1], [0, 0], 'k-')  # Black line between points
    
    # Set grid and ticks
    ax.set_aspect('equal')  # Ensure equal spacing
    plt.grid(False)         # Disable the default grid lines
    plt.xticks(range(n))    # Set x-ticks to the chain points
    plt.yticks([])          # No y-ticks since it's a 1D chain
    plt.xlim(-0.5, n-0.5)   # Limits for x-axis
    plt.ylim(-0.5, 0.5)     # Limits for y-axis to give space around the chain

    plt.title(f"Simple Chain of Length {n}")
    plt.show()

# Test the function
draw_chain(8)
```

### Problem 3 

Draw a complete graph with your own implementation.

### Solution 

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_complete_graph(n):
    
    # Create figure and axis
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Generate points for the vertices positioned in a circular layout (common format for complete graphs)
    angles = np.linspace(0, 2 * np.pi, n, endpoint=False) # discretize the angle space into n points 
    x = np.cos(angles)  # X coordinates of the vertices 
    y = np.sin(angles)  # Y coordinates of the vertices 
    
    # Plot vertices 
    ax.scatter(x, y, s=100, color='b', zorder=2)  # Blue circles for nodes
    
    # Plot edges
    for i in range(n):
        for j in range(i + 1, n):
            ax.plot([x[i], x[j]], [y[i], y[j]], color='gray', zorder=1)  # Gray lines for edges

    # Set equal scaling and remove axis
    ax.set_aspect('equal')
    plt.axis('off')  # Turn off the axes for a cleaner look

    # Add title
    plt.title(f"Complete Graph of size {n}", fontsize=16)
    
    # Show the plot
    plt.show()

# Test the function with n = 5
draw_complete_graph(20)
```

### Problem 4 

Draw a weighted graph.

### Solution 

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_weighted_graph():
    # Number of nodes
    n = 6
    
    # Node positions (circular layout)
    angles = np.linspace(0, 2 * np.pi, n, endpoint=False)
    x = np.cos(angles)  # X coordinates of the nodes
    y = np.sin(angles)  # Y coordinates of the nodes

    # Edges and their weights (list of tuples: (node1, node2, weight))
    edges = [
        (0, 1, 5),  # Edge between node 0 and node 1 with weight 5
        (1, 2, 3),  # Edge between node 1 and node 2 with weight 3
        (2, 3, 7),  # Edge between node 2 and node 3 with weight 7
        (3, 4, 2),  # Edge between node 3 and node 4 with weight 2
        (4, 5, 6),  # Edge between node 4 and node 5 with weight 6
        (5, 0, 4),  # Edge between node 5 and node 0 with weight 4
    ]
    
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Plot vertices 
    ax.scatter(x, y, s=100, color='b', zorder=2)  # Blue circles for nodes
    
    # Plot edges and display weights
    for edge in edges:
        i, j, weight = edge
        ax.plot([x[i], x[j]], [y[i], y[j]], color='gray', zorder=1)  # Gray lines for edges
        # Display the weight in the middle of the edge
        mid_x = (x[i] + x[j]) / 2
        mid_y = (y[i] + y[j]) / 2
        ax.text(mid_x, mid_y, str(weight), color='red', fontsize=12, fontweight='bold')

    # Set equal scaling and remove axis
    ax.set_aspect('equal')
    plt.axis('off')  # Turn off the axes for a cleaner look

    # Add title
    plt.title("Weighted Graph with 6 Vertices", fontsize=16)
    
    # Show the plot
    plt.show()

# Call the function to draw the weighted graph with 6 nodes
draw_weighted_graph()
```

### Problem 5 

Draw a directed graph.

### Solution 

```python
import matplotlib.pyplot as plt
import numpy as np

def draw_weighted_graph():
    # Number of nodes
    n = 6
    
    # Node positions (circular layout)
    angles = np.linspace(0, 2 * np.pi, n, endpoint=False)
    x = np.cos(angles)  # X coordinates of the nodes
    y = np.sin(angles)  # Y coordinates of the nodes

    # Edges and their weights (list of tuples: (node1, node2, weight))
    edges = [
        (0, 1, 5),  # Directed edge from node 0 to node 1 with weight 5
        (1, 2, 3),  # Directed edge from node 1 to node 2 with weight 3
        (2, 3, 7),  # Directed edge from node 2 to node 3 with weight 7
        (3, 4, 2),  # Directed edge from node 3 to node 4 with weight 2
        (4, 5, 6),  # Directed edge from node 4 to node 5 with weight 6
        (5, 0, 4),  # Directed edge from node 5 to node 0 with weight 4
    ]
    
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Plot vertices 
    ax.scatter(x, y, s=100, color='b', zorder=2)  # Blue circles for nodes
    
    # Plot edges, arrows, and display weights
    for edge in edges:
        i, j, weight = edge
        
        # Plot the directed edge (arrow) using annotate function
        ax.annotate("",
                    xy=(x[j], y[j]), xycoords='data',
                    xytext=(x[i], y[i]), textcoords='data',
                    arrowprops=dict(arrowstyle="->", color='gray', lw=2),
                    zorder=1)
        
        # Display the weight in the middle of the edge
        mid_x = (x[i] + x[j]) / 2
        mid_y = (y[i] + y[j]) / 2
        ax.text(mid_x, mid_y, str(weight), color='red', fontsize=12, fontweight='bold')

    # Set equal scaling and remove axis
    ax.set_aspect('equal')
    plt.axis('off')  # Turn off the axes for a cleaner look

    # Add title
    plt.title("Directed Weighted Graph with 6 Vertices", fontsize=16)
    
    # Show the plot
    plt.show()

# Call the function to draw the directed weighted graph with 6 nodes
draw_weighted_graph()
```

### Problem 6 

Compute the vertex degree for one of the graphs above.

### Solution 

I provide the solution using the graph of Problem 3

```python
# size of the graph 
size = 10

# draw a compete graph 
draw_complete_graph(size)

def compute_vertex_degree(n):
    if n < 1:
        raise ValueError("Size of the graph must be at least 1")
    # In a complete graph of size n, the degree of each vertex is n-1
    degree = n - 1
    return degree

degree = compute_vertex_degree(size)
print(f"The vertex degree of a complete graph of size {size} is: {degree}")
```
